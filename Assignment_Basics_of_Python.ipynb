```python
import random, math
S = [random.randint(100,900) for i in range(100)]
odd = []
even = []
prime = []

def prime_or_not(n):
  for i in range(2,math.isqrt(n)+1):
    if n%i == 0:
      return False
  return True

for i in S:
  if i%2 == 0:
    even.append(i)
  else:
    odd.append(i)

for x in S:
  if prime_or_not(x):
    prime.append(x)

print('S = ',S)
print('\nlength = ',len(S))
print('\nodd = ',odd)
print('\nlength = ',len(odd))
print('\neven = ',even)
print('\nlength = ',len(even))
print('\nprime = ',prime)
print('\nlength = ',len(prime))
```

```python
students = {'Ameya': [85, 90, 78, 92, 88], 'John': [70, 65, 72, 68, 74], 'Luke': [95, 98, 94, 97, 96], 'Gloria': [60, 58, 61, 63, 59], 'Rachel': [80, 82, 85, 79, 81]}
average = {}
for name, marks in students.items():
    average[name] = sum(marks) / len(marks)
print(average)
avglist = list(average.values())
maxavg = max(avglist)
minavg = min(avglist)

for name, avg in average.items():
    if avg == maxavg:
        print('Maximum average -', name, ':', maxavg)
    if avg == minavg:
        print('Minimum average -', name, ':', minavg)
```
